# EDA & Feature Engineering Flight Price Prediction

# Features
The various features of the cleaned dataset are explained below:<br>
<br>
1.Airline:The name of the airline company is stored in the airline column.It is categorical feature having 6 different airlines.<br>
<br>
2.Flight:Flight stores information regarding the plane's flight code. It is a categorical feature.<br>
<br>
3.Source City:City from which the flight takes off. It is a categorical feature having 6 unique cities.<br>
<br>
4.Department Time: This is a derived categorical feature obtained created by grouping time periods into bins.It stores information about the departure time and have 6 unique time labels.<br>
<br>
5.Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities.<br>
<br>
6.Arrival Time: This is derived categorical feature created by grouping time intervals into bins. It has six distinct labels information about the arrival time<br>
<br>
7.Destination City: City where the flight will land.It is a categorical feature having 6 unique cities.<br>
<br>
8.Class: A categorical feature that contains informatuion on seat class; it has two distict values:Business and Economy.<br>
<br>
9.Duration: A continuous feature that displays the overall amount of time it takes to travel between cities in hours.<br>
<br>
10.Days Left: This is a derived characteristic that is calculated by subtracting the trip date by the booking date.<br>
<br>
11.Price: Target variable stores information of the ticket price.<br>

## Feature Engineering

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
# Converting the datatype into int
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
## Drop Date of Journey
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

# Time splitting

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```
